```python

from sklearn.cluster import KMeans
from sklearn import preprocessing
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
dataframe=pd.read_csv("/content/kMEAN_CLUSTERING/Mall_Customers.csv")
dataframe.head()
```

```python
dataframe.shape
```

```python
scaler=MinMaxScaler()
scale=scaler.fit_transform(dataframe[['Annual Income (k$)','Spending Score (1-100)']])
df_scale=pd.DataFrame(scale,columns=['Annual Income (k$)','Spending Score (1-100)']);
df_scale.head()
```

```python
km=KMeans(n_clusters=2)
y_predicted=km.fit_transform(df_scale[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

```python
km.cluster_centers_
```

```python
dataframe['cluster']=km.labels_
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue='cluster',data=dataframe,palette='viridis')
```

```python
K=range(2,12)
wss=[]
for  k in K:
  kmeans=cluster.KMeans(n_clusters=k)
  Kmeans=kmeans.fit(df_scale)
  wss_iter=kmeans.inertia_
  wss.append(wss_iter)
```

```python
plt.xlabel('K')
plt.ylabel('wss')
plt.plot(K,wss)
```

```python
km=KMeans(n_clusters=5)
y_prediction=km.fit_transform(df_scale,[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```
